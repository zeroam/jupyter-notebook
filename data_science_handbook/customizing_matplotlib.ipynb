```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np

%matplotlib inline
```

```python
x = np.random.randn(1000)
plt.hist(x);
```

```python
# use a gray background
ax = plt.axes(facecolor='#E6E6E6')
ax.set_axisbelow(True)

# draw sloid white grid lines
plt.grid(color='w', linestyle='solid')

# hide axis splines
for spline in ax.spines.values():
    spline.set_visible(False)
    
# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# ligten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
# control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

# Changing the Defaults: `rcParams`
- runtime configuration(rc)

```python
IPython_default = plt.rcParams.copy()
```

```python
from matplotlib import cycler
colors = cycler('color', ['#EE6666', '#3388BB', '#9988DD', '#EECC55', '#88BB44', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none', axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.randn(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

## FiveThirtyEight style
- `fivethirtyeight` style mimics the graphics found on the popular [FiveThirtyEight website](https://fivethirtyeight.com/)

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

## ggplot
- The `ggplot` package in the R language is a very popular visualization tool

```python
with plt.style.context('ggplot'):
    hist_and_lines()
```
